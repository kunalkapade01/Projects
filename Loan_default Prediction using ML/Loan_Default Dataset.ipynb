```python
# Resampling:- Oversampling

from imblearn.over_sampling import RandomOverSampler
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_sampled, y_sampled, test_size=0.25, random_state=12)
```

```python
# Convert the data into standard scaler:--
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x_tr_sc = scaler.fit_transform(x_train)
x_te_sc = scaler.transform(x_test)
```

```python
x_tr_sc
```

```python
x_te_sc
```

# Apply the ML model:

- 𝐑𝐚𝐧𝐝𝐨𝐦 𝐅𝐨𝐫𝐞𝐬𝐭:-

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()
```

```python
model.fit(x_tr_sc, y_train)
```

```python
model.score(x_tr_sc, y_train)
```

```python
pred = model.predict(x_te_sc)
pred
```

```python
#find accuracy:
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
accuracy = accuracy_score(y_test, pred)
print(f"𝐀𝐜𝐜𝐮𝐫𝐚𝐜𝐲 𝐒𝐜𝐨𝐫𝐞: {accuracy}")
```

```python
conf_matrix = confusion_matrix(y_test, pred)
print(f"𝐂𝐨𝐧𝐟𝐮𝐬𝐢𝐨𝐧 𝐌𝐚𝐭𝐫𝐢𝐱: \n {conf_matrix}")
```

```python
class_report = classification_report(y_test, pred)
print(f"𝐂𝐥𝐚𝐬𝐬𝐢𝐟𝐢𝐜𝐚𝐭𝐢𝐨𝐧 𝐑𝐞𝐩𝐨𝐫𝐭: \n {class_report}")
```
